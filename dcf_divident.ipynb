```python
import numpy as np
import pandas as pd
import yfinance as yf

#data = [{"year":"2024","dividend":0.3},{"year":"2023","dividend":0.2},{"year":"2023","dividend":0.35},{"year":"2022","dividend":0.25},{"year":"2022","dividend":0.25},{"year":"2021","dividend":0.25},{"year":"2021","dividend":0.3},{"year":"2020","dividend":0.2},{"year":"2020","dividend":0.4},{"year":"2019","dividend":0.3},{"year":"2019","dividend":0.35}]

# Dividend data from the user input
#dividends = [sum(item["dividend"] for item in data if item["year"] == year) for year in sorted(set(item["year"] for item in data), reverse=True)]

#print(dividends)

def calculate_value(dividends):
    # Years corresponding to the dividends
    years = np.arange(1, len(dividends) + 1)

    # Assuming a discount rate of 8% (r = 0.08)
    r = 0.08

    # Calculate the present value of dividends
    pv_dividends = dividends / (1 + r) ** years

    # Estimate the growth rate (CAGR)
    initial_dividend = dividends[-1]
    final_dividend = dividends[0]
    years_between = len(dividends) - 1

    g = (final_dividend / initial_dividend) ** (1 / years_between) - 1

    # Calculate the terminal value using the last dividend and growth rate
    terminal_value = dividends[0] * (1 + g) / (r - g)

    # Calculate the present value of the terminal value
    pv_terminal_value = terminal_value / (1 + r) ** len(dividends)

    # Sum the present values of the dividends and the terminal value to get the intrinsic value
    intrinsic_value = np.sum(pv_dividends) + pv_terminal_value
    #print(intrinsic_value, g)
    return intrinsic_value

#calculate_value([0.93, 1.03, 1.04, 1.04, 1.03])

df = pd.read_csv('stock_dividends.csv')

calculated_values = []
current_price_values = []

# Iterate through the DataFrame
for index, row in df.iterrows():
    data = [row['2023'], row['2022'], row['2021'], row['2020'], row['2019']]
    
    # Calculate the value and append it to the list
    result = calculate_value(data)
    calculated_values.append(result)

    #print(row['Stock'] + '.BKK')
    stock = yf.Ticker(row['Stock'] + '.BK')
    current_price = stock.history(period="1d")['Close'].iloc[-1]
    current_price_values.append(current_price)

# Assign the list as a new column in the DataFrame
df['CalculatedValue'] = calculated_values
df['CurrentPriceValue'] = current_price_values
# Print the updated DataFrame to verify
print(df)


# Example: Get the current price for SCB (Siam Commercial Bank)
ticker = "SCB.BK"
stock = yf.Ticker(ticker)

# Get the latest market price
current_price = stock.history(period="1d")['Close'].iloc[-1]

print(f"Current price of {ticker}: {current_price} THB")
```
